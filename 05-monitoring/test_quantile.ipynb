```python
import pandas as pd

# Load the parquet file
file_path = '/workspaces/MLOps-Zoomcamp/05-monitoring/green_tripdata_2024-03.parquet'
df = pd.read_parquet(file_path)

```

```python
df.tail()
```

```python
df['lpep_pickup_datetime'] = pd.to_datetime(df['lpep_pickup_datetime'])

# Crear una lista para almacenar los quantiles diarios
quantile_list = []
dates_list = []

# Crear un rango de fechas desde el 1 de marzo hasta el 31 de marzo
start_date = pd.to_datetime('2024-03-01')
end_date = pd.to_datetime('2024-03-31')

for single_date in pd.date_range(start_date, end_date):
    # Filtrar los datos para el día actual
    daily_data = df[df['lpep_pickup_datetime'].dt.date == single_date.date()]
    
    if not daily_data.empty:
        # Calcular el quantile 0.5 para 'fare_amount' en el día actual
        daily_quantile = daily_data['fare_amount'].quantile(0.5)
        
        # Guardar el resultado en la lista
        quantile_list.append(daily_quantile)
        dates_list.append(single_date.date())

# Crear un DataFrame con los resultados diarios
daily_quantile_df = pd.DataFrame({
    'date': dates_list,
    'quantile_0.5': quantile_list
})

# Encontrar el valor máximo de la lista de quantiles diarios
max_daily_quantile = daily_quantile_df['quantile_0.5'].max()

print(daily_quantile_df)
print(f"La mediana diaria máxima (quantile = 0.5) para 'fare_amount' en marzo de 2024 es {max_daily_quantile}")
```
